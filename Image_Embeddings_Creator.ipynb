```
import keras
from tqdm import tqdm
```

```
from keras.applications import ResNet50
```

```
model = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3),pooling='avg')
```

```
model.summary()
```

```
!wget 'http://nlp.cs.illinois.edu/HockenmaierGroup/Framing_Image_Description/Flickr8k_Dataset.zip'
```

```
from keras.preprocessing import image
```

```
with open('./Flickr_8k.devImages.txt') as f: #do similarly for test and train.
    trainer = f.readlines()
```

```
train = [s[:-1] for s in trainer]
```

```
len(train)
```

```
path = './Flicker8k_Dataset/'
imgData = {}
for ix in tqdm(range(len(train))):
    imPath = path+train[ix]
    img = image.load_img(imPath,target_size=(224,224))
    img_array = image.img_to_array(img)
    img_array = img_array.reshape((1,224,224,3))
    out = model.predict(img_array)
    imgData[train[ix]] = out 
```

```
import pickle
```

```
with open('val_embs.pickle', 'wb') as handle:
    pickle.dump(imgData, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
